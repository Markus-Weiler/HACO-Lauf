```python
# Der reguläre Ausdruck
pattern = r"(\d{1,2}:\d{2}:\d{2})|(\d{1,2}:\d{2})"
```

```python
def find_time_formats_in_string(input_string):
    # Der reguläre Ausdruck
    pattern = r"(\d{1,2}:\d{2}(?::\d{2})?)"

    # Verwendung von re.findall, um alle Vorkommen zu finden
    matches = re.findall(pattern, input_string)

    # Ausgabe der gefundenen Zeiten
    return matches


print(find_time_formats_in_string('01:00:39,6'))
print(find_time_formats_in_string('58:23,3'))
```
